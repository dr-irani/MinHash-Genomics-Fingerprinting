```python
df = pd.read_csv('synth_trueJ.csv', header=None)
column_names = ['trueED']
for k in [4, 8, 12, 16, 20]:
    for l in [32, 64, 128, 256, 512]:
        column_names.append('[' + str(k) + ',' + str(l) + ']')
df.columns = column_names
fig, ax = plt.subplots(1,1,figsize=figsize)
lines = df.iloc[:, 1:]
ax.plot(df.loc[:, 'trueED'],  lines, linestyle='None', marker='.', markersize=15.0)
ax.set(xlabel='True Edit Distance', ylabel='True Jaccard')
ax.legend(lines)
plt.savefig('jacedit.png')
```
